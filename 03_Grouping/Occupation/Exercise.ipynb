### Step 1. Import the necessary libraries

```python
import pandas as pd 
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
```

### Step 4. Discover what is the mean age per occupation

```python
users.head()
```

```python
users.groupby("occupation")["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users_occ = users.groupby("occupation")
```

```python
occupation_gender_counts = users.groupby('occupation')['gender'].value_counts()

# Create a new DataFrame with occupation and gender counts
occupation_gender_df = occupation_gender_counts.unstack(fill_value=0)

# Calculate the ratio of M to F for each occupation
occupation_gender_df['ratio_m_to_f'] = occupation_gender_df['M'] / occupation_gender_df['F']

# Print the resulting DataFrame
occupation_gender_df[['M', 'F', 'ratio_m_to_f']]
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users_occ["age"].agg(["min", "max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])["age"].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
occ_gender_counts = users_occ["gender"].value_counts()
occ_gender_unstack = occ_gender_counts.unstack(fill_value=0)
occ_gender_unstack["total"] = occ_gender_unstack["F"] + occ_gender_unstack["M"]
occ_gender_unstack["f_percent"] = (occ_gender_unstack["F"] / occ_gender_unstack["total"]) * 100
occ_gender_unstack["m_percent"] = (occ_gender_unstack["M"] / occ_gender_unstack["total"]) * 100
occ_gender_unstack
```
